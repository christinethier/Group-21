````python
# YOUR CODE HERE
import pandas as pd
import re

def process_weather(df):

    '''
    This function process the given dataframe by structuring and cleaning it.
    - Structuring includes removing unused columns, renaming and selecting only observations 
    of maximum temperature. 
    - Cleaning includes inserting missing decimal, sorting and resetting index.
    '''

    # structure and clean data using methods chaining
    # note that the original columns now are strings when loading the csv file
    # and not integers as when downloading the data
    
    df_out = \
        df\
            .rename(columns={'0': 'station', '1': 'datetime', '2': 'obs_type', '3': 'obs_value'})\
            .query("obs_type == 'TMAX'")\
            .assign(obs_value=lambda df: df['obs_value']/10)\
            .sort_values(by=['datetime','station'])\
            .reset_index(drop=True)\
            .copy() 
    # In the above we changed the sorting order to sorting by datetime first
    
    # Make area code column
    area_codes = []
    for x in df_out['station']:
        area_codes.append("".join(re.findall(r"[a-zA-Z]+", x)))
    df_out['area_code'] = area_codes
    df_out['area_code'] = df_out['area_code'].astype('category')
    
    # Transform to and make new datetime columns
    df_out['datetime_dt'] = pd.to_datetime(df_out['datetime'], format = '%Y%m%d')
    df_out['month'] = df_out['datetime_dt'].dt.month
    df_out['year'] = df_out['datetime_dt'].dt.year    
    
    return df_out

# Loading the weather data:
weather_data = pd.read_csv('weather_data_1870-1875.csv')

# Running the function with the weather data and saving the processed data frame as a variable
df_weather_period = process_weather(weather_data)
display(df_weather_period)
````

````python
assert any(df_weather_period.station == process_weather(pd.read_csv('weather_data_1870-1875.csv')).station)
assert any(df_weather_period.columns.values == ['station', 'datetime', 'obs_type', 'area', 
                                                'obs_value', 'datetime_dt', 'month', 'year'])
assert len(df_weather_period) == 132317
assert round(df_weather_period.obs_value.mean()) == 14
````
